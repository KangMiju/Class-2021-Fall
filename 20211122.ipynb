```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds
freq = 100.0     # sine frequency, Hz
phase = np.pi/2  # phase, radian
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

D = 293.66
E = 329.63
F = 349.23
G = 392
A = 440
B =493.88
C = 523.25
```

```python
#Christmas Carol medley

notes = [G, A, G, E, 0, G, A, G, E, 0, D*2, D*2, B, 0, C, C, G, 0, 
         A, A, C, B, A, G, A, G, E, 0, A, A, C, B, A, G, A, G, E, 0, 
         D*2, D*2, F*2, D*2, B, C, E*2, 0, C*2, G*2, E*2, G*2, F*2, D*2, C]

durs = [0.75, 0.25, 0.5, 1.0, 0.5, 0.75, 0.25, 0.5, 1.0, 0.5, 1, 0.5, 1, 0.5, 1, 0.5, 1, 0.5, 1, 0.5,
        1, 0.5, 0.5, 0.5, 0.5, 0.75, 0.25, 0.5, 1, 0.5, 1, 0.5, 0.5, 0.5, 0.5, 0.75, 0.25, 0.5, 1, 0.5, 
        1, 0.5, 0.5, 0.5, 0.5, 1.5, 1, 0.5, 0.5, 0.5, 0.75, 0.25, 0.5, 0.25]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
